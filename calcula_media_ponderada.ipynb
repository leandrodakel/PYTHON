```python
peso1 = float(input("Digite o peso da nota 01:"))
peso2 = float(input("Digite o peso da nota 02:"))
print(f"\n")

class Alunos:
  def __init__(self,nome,turma,nota1,nota2):
    self.nome = nome
    self.turma = turma
    self.nota1 = nota1
    self.nota2 = nota2

  def media(self):
    soma_produtos = (self.nota1 * peso1) + (self.nota2 * peso2)
    soma_pesos = peso1 + peso2
    total1 = soma_produtos / soma_pesos
    return total1

  def media_final(self):
    soma_produtos = (self.nota1 * peso1) + (self.nota2 * peso2) + (self.nota3 * peso3)
    soma_pesos = peso1 + peso2 + peso3
    total2 = soma_produtos / soma_pesos
    return total2

  def nota_final(self):
    return (self.nota1 * peso1) + (self.nota2 * peso2) + (self.nota3 * peso3) / (peso1 + peso2 + peso3)

  def status(self):
    return f"Aluno: {self.nome}\n Turma: {self.turma}\n Notas: {self.nota1}, {self.nota2}\n"


lista_alunos = []

n = 0
i = 0

n = int(input("Número de alunos a serem cadastrados:"))
print(f"\n")

while i < n:
  print(f"Digite os dados do aluno 0{i+1}:")
  print(f"\n")
  if i >= 9:
    print(f"Digite os dados do aluno {i+1}:")
    print(f"\n")
  nome = input("Nome:")
  turma = input(" Turma:")
  nota1 = float(input(" Nota 01:"))
  nota2 = float(input(" Nota 02:"))
  print(f"\n")
  aluno = Alunos(nome,turma,nota1,nota2)
  lista_alunos.append(aluno)
  print(f"\n")
  i += 1

k = 0

while k != 2:
  encontrar_aluno = input("Digite o nome do aluno para calcular sua média: ")
  print(f"\n")
  encontrado = False

  for aluno in lista_alunos:
    if aluno.nome == encontrar_aluno:
      if aluno.media() >= 6:
        print(aluno.status())
        print(f"A media do aluno '{aluno.nome}' é: {aluno.media()}")
        print(f"\n")
        print(f"O aluno '{aluno.nome}' está APROVADO")
        print(f"\n")
      else:
        print(aluno.status())
        print(f"A media do aluno '{aluno.nome}' é: {aluno.media()}")
        print(f"\n")
        print(f"O aluno '{aluno.nome}' está em RECUPERAÇÃO")
        print(f"\n")
        peso3 = float(input("Digite o peso da nota 03:"))
        nota3 = float(input("Digite a nota 03 do aluno:"))
        print(f"\n")
        print(f"A media final do aluno '{aluno.nome}' é: {aluno.media_final()}")
        if aluno.media_final() >= 6:
          print(f"O aluno '{aluno.nome}' está APROVADO")
          print(f"\n")
        else:
          print(f"O aluno '{aluno.nome}' está REPROVADO")
          print(f"\n")

      encontrado = True
      break

  if not encontrado:
    print("Aluno não cadastrado")
    print(f"\n")

  k = int(input("1- Continuar | 2 - Sair "))
  print(f"\n")
```

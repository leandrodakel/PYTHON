```python
# Importando a biblioteca Matplotlib (para o gráfico requerido)
import matplotlib.pyplot as plt

# Definindo uma classe
class Livros:
  def __init__(self, titulo, autor, genero, quantidade):
    self.titulo = titulo
    self.autor = autor
    self.genero = genero
    self.quantidade = quantidade

  # Criando lista vazia para armazenar os dados dos livros
  lista_de_livros = []

  # Estabelecendo uma função para cadastro de livros
  def cadastrar_livro(self):
    operador_cadastro = 0
    while operador_cadastro != 2:
      print("Cadastro de livro")
      print("------------------------------------------------")

      titulo = input("\nDigite o título do livro: ")
      autor = input("Digite o autor do livro: ")
      genero = input("Digite o gênero do livro: ")
      quantidade = int(input("Digite a quantidade de livros: "))
      livro = Livros(titulo,autor,genero,quantidade)
      self.lista_de_livros.append(livro)
      print("Livro cadastrado com sucesso!\n")
      print("------------------------------------------------")

      operador_cadastro = int(input("1 - Continuar | 2 - Voltar ao MENU "))
      print("------------------------------------------------\n")

  # Criando uma função para exibir a lista de livros cadastrados
  def listar_livros(self):
    print("Lista de livros")
    print("------------------------------------------------\n")
    for livro in self.lista_de_livros:
      print(f"{livro.titulo} (Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade})\n")
    print("------------------------------------------------\n")

  # Criando uma função de busca de livros
  def buscar_livro(self):
    print("Busca de livros")
    print("------------------------------------------------\n")

    # Cirando uma variável de controle para o laço de repetição de busca
    operador_busca = 0

  # Estruturando uma repetição do tipo "do while" (permite várias buscas)
    while operador_busca != 2:

      # Fazendo a leitura do título desejado
      encontrar_titulo = input("Digite o título do livro: ")

      # Declarando variável booleana para livros inexistentes no cadastro
      encontrado = False

      # Incluindo livros na lista de livros por meio de um laço de repetição
      print("\n")
      for livro in self.lista_de_livros:

        # Constatando se o livro é existente no cadastro
        if livro.titulo == encontrar_titulo:
          print(f"{livro.titulo} (Autor: {livro.autor}, Gênero: {livro.genero}, Quantidade: {livro.quantidade})\n")
          print("------------------------------------------------")
          # Alterando a variável de controle (no caso de livro encontrado)
          encontrado = True

      # Constatando se o livro é existente no cadastro
      if not encontrado:
        print("Livro não encontrado!\n")
        print("------------------------------------------------")

      # Permitindo ao usuário fazer uma nova busca
      operador_busca = int(input("1 - Continuar | 2 - Voltar ao MENU "))
      print("------------------------------------------------\n")

  # Criando uma função para exibir um gráfico de gêneros x quatidades de livros
  def gerar_grafico(self):

    # Criando listas vazias para os eixos x e y do gráfico
    generos = []
    quantidades = []

    # Transportando informções da lista de livros para as listas novas
    for livro in self.lista_de_livros:
      generos.append(livro.genero)
      quantidades.append(livro.quantidade)

    # Gerando um grafico de barras
    plt.bar(generos, quantidades)

    # Adicionando rótulos aos eixos
    plt.xlabel('Gêneros')
    plt.ylabel('Quantidade disponível')

    # Adicionando um título ao gráfico
    plt.title('GÊNEROS DE LIVROS POR QUANTIDADE')

    # Mostrando o gráfico
    plt.show()

    print("------------------------------------------------")

print("------------------------------------------------")
print("Bem-vindo ao Gerenciador de livros da Biblioteca")
print("------------------------------------------------\n")

# Declarando variável de controle do menu geral do programa
menu = 0

# Criando looping para as opções
while menu != 5:
  print(f"1 - Cadastrar livro\n2 - Listar livros\n3 - Buscar livro\n4 - Gerar gráfico (gênero x quantidade)\n5 - Sair")

  # Lendo a opção do usuário
  menu = int(input())
  print("\n")

  print("------------------------------------------------")

  # Acionando a função de cadastro de livros
  if menu == 1:
    livro = Livros("","","",0)
    livro.cadastrar_livro()

  # Acionando a função para listar livros
  elif menu == 2:
    livro = Livros("","","",0)
    livro.listar_livros()

  # Acionando a função de busca de livros
  elif menu == 3:
    livro = Livros("","","",0)
    livro.buscar_livro()

  # Acionando a função para gerar o gráfico
  elif menu == 4:
    livro = Livros("","","",0)
    livro.gerar_grafico()
    print("------------------------------------------------\n")

  # Validadndo a saída do loop
  elif menu == 5:
    print("Saindo...")
    print("------------------------------------------------\n")

  # Avisando que a opção não é valida
  else:
    print("Opção inválida!")
    print("------------------------------------------------\n")
```
